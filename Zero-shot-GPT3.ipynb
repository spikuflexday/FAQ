```python
import openai
def get_questions(context):
    try:
        response = openai.Completion.create(
            engine="davinci-instruct-beta-v2",
            prompt=f"Write questions based on the text below\n\nText: {context}\n\nQuestions:\n1.",
            temperature=0,
            max_tokens=257,
            top_p=1,
            frequency_penalty=0,
            presence_penalty=0,
            stop=["\n\n"]
        )
        return response['choices'][0]['text']
    except:
        return ""
openai.api_key = 'ENTER_YOUR_KEY'
```

```python
data2 = {}
with open("/Users/siddhartha/Documents/FAQ dataset/"+"faq-dataset-doc2.txt") as file:
    if '/'*10 in file.readline():
        context = file.readline()
    else:
        context = ''
    while context:
        j = file.readline()
        while True:
            #print('context: ',j)
            if '/'*10 in j:
                break
            context += j
            j = file.readline()
        #print(context)
        qa = []
        flag = 0
        while flag == 0:
            q = ''
            j = file.readline()
            #print('q: ',j)
            while True:
                #print('q: ',j)
                if '/'*2 in j:
                    break
                q += j
                j = file.readline()
            #print(q)
            a = ''
            j = file.readline()
            while True:
                if '/'*2 in j:
                    if '/'*10 in j:
                        flag = 1
                    break
                a += j
                j = file.readline()
            qa.append((q,a))
        data2[context] = qa
        context = file.readline()
```
